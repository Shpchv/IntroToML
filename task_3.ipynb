```python
import pandas as pd 
from scipy.spatial import distance
import numpy as np
```

```python
data = pd.read_csv ('/content/data.csv', delimiter=';', index_col='id')
data
```

```python
x = np.array([68,46])
```

```python
data['Euclidean'] = [distance.euclidean(obj, x) for index, obj in data.iloc[:,:2].iterrows()]
data
```

```python
data['Manhattan'] = [distance.cityblock(obj, x) for index, obj in data.iloc[:,:2].iterrows()]
data
```

```python
data.style.highlight_min(['Euclidean', 'Manhattan'], color = 'darkgreen', axis = 0)
```
